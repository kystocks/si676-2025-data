## Modifying our requests with parameters

```python
parameters = {'fo':'json'}
```

## try a search

```python
search_endpoint = '/search'
```

```python
parameters = {'fo':'json', 'q':'yarn'}
```

```python
r = requests.get(url + search_endpoint, params = parameters)
```

```python
print(r.url, r.status_code)
```

```python
response = r.json()

for info in response:
    print(info)
```
